```python
def salary_bucket(x):
    if pd.isna(x):
        return "Unknown"
    elif x < 50000:
        return "Low"
    elif x < 120000:
        return "Mid"
    else:
        return "High"

df2["SalaryBand"] = df2["ConvertedCompYearly"].apply(salary_bucket)

print(df2["SalaryBand"].value_counts())
```

```python
print(df2["AISelect"].value_counts())
print(df2["AIAcc"].value_counts())
print(df2["WorkExp"].describe())
```

```python
def exp_cohort(x):
    if pd.isna(x):
        return None
    elif x <= 4:
        return "Junior (0-4)"
    elif 10 <= x <= 15:
        return "Senior (10-15)"
    else:
        return None   # anyone outside these two bands, we ignore for this comparison

df2["ExpCohort"] = df2["WorkExp"].apply(exp_cohort)
```

```python
cohort_df = df2[df2["ExpCohort"].notna()]
```

```python
cohort_df.groupby(["ExpCohort", "AISelect"]).size()
```

```python
cohort_df.groupby(["ExpCohort", "AIAcc"]).size()
```
